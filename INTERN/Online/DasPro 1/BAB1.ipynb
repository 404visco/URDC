```python
#FUNGSI
def tambah(a,b):
    return a+b
def kurang(a,b):
    return a-b
def luas_lingkaran(jari2):
    return 3.14*(jari2**2)
def is_even(a):
    return a%2==0
def fibonacci(a): #rekursif
    if a==0:
        return 0
    if a==1:
        return 1
    else:
        return fibonacci(a-1)+fibonacci(a-2)

print(tambah(1,2))
print(kurang(3,4))
print(luas_lingkaran(7))
print(is_even(3))
for i in range(5):
    print(fibonacci(i))
```

```python
#Kelas
class Robot:
    pass

robot_tempur = Robot()

class Kucing:
    def __init__(self, nama, warna):
        self.nama = nama
        self.warna = warna

    def deskripsi(self):
        print(f"Nama Kucingnya {self.nama}, Warnanya {self.warna}")
        
print(Kucing("Yogi", "Oren").nama)
print(Kucing("Yogi", "Oren").warna)
print(Kucing("Yogi", "Oren").deskripsi())

class Luas:
    def __init__(self, sisi, radius):
        self.sisi=sisi
        self.radius=radius

    def Persegi(self, Lebar=1):
        if Lebar == 1:
            return self.sisi*self.sisi
        else:
            return self.sisi*Lebar
    
    def Lingkaran(self):
        return 3.14*(self.radius**2)
    
print(Luas(10,7).Persegi())
print(Luas(10,7).Persegi(11))
print(Luas(10,7).Lingkaran())
```

```python
#PENERAPAN PLOT (NUMPY+MATPLOTLIB)
import numpy as np
import matplotlib.pyplot as plt

#REGRESI
X= np.array([1,2,3,4,5])
Y= np.array([1,4,9,16,25])

#Y= a+bX
koef = np.polyfit(X,Y,1)
b, a = koef

#predicted values
Y_pred = a+b*X

#Representasi plot
plt.scatter(X, Y) #--> Menampilkan persebaran X dan Y
plt.plot(X, Y_pred) #--> Buat Grafik
plt.title("regresi linear") #--> Judul
plt.show() #--> Menampilkan plot
```
