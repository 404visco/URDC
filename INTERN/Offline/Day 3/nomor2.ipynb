```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
confusion_matrix = np.array([
    [120, 8, 2, 1, 0],
    [6, 95, 12, 4, 3],
    [3, 10, 150, 15, 5],
    [2, 5, 20, 100, 10],
    [1, 2, 5, 12, 80]
])
kelas = ['Apel', 'Jeruk', 'Pisang', 'Mangga', 'Anggur']
```

```python
#a. Total akurasi model
total_benar = 0
for i in range(len(confusion_matrix)):
    total_benar+=confusion_matrix[i][i]
jumlahdata = np.sum(confusion_matrix)
print(f'Total Akurasi: {total_benar/jumlahdata*100}%')
```

```python
#B. Presisi untuk setiap kelas
#Nilai prediksi Benar / Total Nilai prediksi

for i in range(len(kelas)):
    total = 0
    for j in range(len(confusion_matrix)):
        total+=confusion_matrix[j][i]
    print(f'Presisi Kelas{i}: {confusion_matrix[i][i]/total*100}%')
```
