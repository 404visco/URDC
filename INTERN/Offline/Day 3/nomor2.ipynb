```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
confussion_matrix = np.array([
    [120, 8, 2, 1, 0],
    [6, 95, 12, 4, 3],
    [3, 10, 150, 15, 5],
    [2, 5, 20, 100, 10],
    [1, 2, 5, 12, 80]
])
kelas = ['Apel', 'Jeruk', 'Pisang', 'Mangga', 'Anggur']
```

```python
#a. Total akurasi model
total_benar = 0
for i in range(len(confussion_matrix)):
    total_benar+=confussion_matrix[i][i]
jumlahdata = np.sum(confussion_matrix)
print(f'Total Akurasi: {total_benar/jumlahdata*100}%')
```

```python
#B. Presisi untuk setiap kelas
#Nilai prediksi Benar / Total Nilai prediksi
presisi_terbaik = 0
kelas_terbaik = ""

for i in range(len(kelas)):
    total = 0
    for j in range(len(confussion_matrix)):
        total+=confussion_matrix[j][i]
    presisi = confussion_matrix[i][i]/total*100
    print(f'Presisi Kelas {kelas[i]}: {presisi}%')

    if presisi > presisi_terbaik:
        presisi_terbaik = presisi
        kelas_terbaik = kelas[i]
```

```python
#C. Kelas dengan Presisi Tertinggi
print(f'Kelas dengan presisi tertinggi: {kelas_terbaik}')
```

```python
#D. 
plt.figure()
plt.imshow(confussion_matrix, cmap= "PuBu")
plt.xticks(range(len(confussion_matrix)), kelas)
plt.yticks(range(len(confussion_matrix)), kelas)
plt.colorbar()
plt.xlabel("Prediksi")
plt.ylabel("Aktual")

for i in range(5):
    for j in range(5):
        plt.text(j, i, confussion_matrix[i, j], ha="center", va="center")

plt.title("Confussion Matrix")
plt.show()
```
