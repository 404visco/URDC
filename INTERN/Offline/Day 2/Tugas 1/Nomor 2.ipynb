```python
import cv2
import numpy as np

# Membaca gambar
image = cv2.imread('gambar.jpg') # BGR format
gray_image = cv2.imread('gambar.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
print('--- Informasi Citra ---')
# Dimensi gambar
height, width = image.shape[:2] # Untuk warna
# height, width = gray_image.shape # Untuk grayscale

# Tipe data
print(f"Dimensi: {width}x{height}")
print(f"Tipe data: {image.dtype}")
print(f"Range nilai: {image.min()} - {image.max()}")
# Untuk gambar warna
channels = image.shape[2] # Jumlah channel
print(f"Jumlah channel: {channels}")
```

```python
print('Modifikasi Pixel (blurring)')
import numpy as np
from scipy.signal import convolve2d
f= np.array(image)
g= np.array([[(1/9) for _ in range(3)] for _ in range(3)])

def konvolusi(matriksinput, matriksfilter):
    dim_line_filter = len(matriksfilter)
    dim_col_filter = len(matriksfilter[0]) 
    dim_line_input = len(matriksinput)
    dim_col_input = len(matriksinput[0])

    dim_line_output = dim_line_input - dim_line_filter+1
    dim_col_output = dim_col_input - dim_col_filter+1

    output= [[0 for _ in range(dim_col_output)]
             for _ in range(dim_line_output)]
    
    for a in range(dim_line_output):
        for b in range(dim_col_output):
            nilai= 0
            for i in range(a,a+dim_line_filter):#Baris
                for j in range(b,b+dim_col_filter): #kolom
                    nilai += matriksinput[i][j]*matriksfilter[i-a][j-b]
            output[a][b]= nilai

    return np.array(output)

print(konvolusi(f,g))
```

```python
print('--- Modifikasi Pixel ---')
image[0,0]= [255,0,0]
image[100,150] = [0,0,255]
cv2.imwrite("modifikasi.jpg", image)
```
